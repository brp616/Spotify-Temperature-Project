```python
import pandas as pd
import numpy as np
datafile = 'GlobalLandTemperaturesByCountry.csv'#Read csv file using pandas
data = pd.read_csv(datafile, names = ['dt', 'Avg Temp', 'Avg Temp Uncertainty', 'Country'], skiprows = 1)
data['dt'] = pd.to_datetime(data['dt'])
# We convert the values to datetime values, so they are easier to manipulate
data['year'] = data['dt'].apply(lambda x: x.year)
# We are only looking at data from the year 2013
data = data[data['year'] == 2013]
data.head()
```
